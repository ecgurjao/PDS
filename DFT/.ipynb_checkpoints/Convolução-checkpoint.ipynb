## Circular shift

The time-shift property of the Fourier transform states that if $X(e^{j\omega})$ the Fourier transform of $x(n)$ then $e^{-j\omega n}X(e^{j\omega n})$ is the transform of  $x[n-m]$. Figure~\ref{fig:shiftsignal} shows that for  Example~\ref{ex:portadiscreta} the sequence obtained by taking the inverse transform of the $N = 16$ DFT points $X(k)$ is periodic. Thus, the product  $e^{-j\frac{2\pi k}{N}m}X(k)$ produces a circular shift in $x(n)$, i.e.  $x_1(n) = x[(n-m)\mod N]$.

### Circular convolution

Linear convolution between a $L$ point signal $x(n)$ and a $M$ point signal $h(n)$ produces a $y(n) = x(n) *h(n)$ with $L + M -1$  points. In the frequency domain convolution corresponds to a multiplication. Taking two $N$ points DFTs $X(k)$ and $H(k)$  calculating $Y(k) = X(k)H(k)$ and applying the inverse transform a $N$ points signal is obtained. That is because that $y(n)$ corresponds to the circular convolution \index{Circular Convolution} of $x(n)$ and $h(n)$, expressed  as 
\begin{equation}
y(n)  = \sum\limits_{k=0}^{N-1}x(k)h[(n-k) \mod N].
\end{equation}

If the DFTs $X(k)$ and $H(k)$ are both size $N$ ($N \geq L, M$) the product $Y(k) = X(k)H(k)$ as well as its inverse DFT  are also of size $N$. In order to obtain  the desired $y(n)$ of length $L + M - 1$, as expected from linear convolution of $x(n)$ and $h(n)$, the DFTs have to be calculated with $L + M - 1$ points. Thus, the product $Y(k)$ will be of length $L + M - 1$ and its inverse DFT is  $y(n)=x(n)*h(n)$, also of length $L+M-1$.

Calculating the DFTs with $L+M-1$ points requires zero padding to increase both $x(n)$ and $h(n)$ to length $L+M-1$, as shown in the next examples.

%y(n)  = \sum\limits_{k=0}^{N-1}x(k)h[(n-k)\mod N].

\section{Overlapp-add method}

%To obtain a linear convolution signals $h(n)$ with length $M$ and $h(n)$ with $L$ points we have to set their lengths, by zero padding, to  $N = M + L -1$. 

Convolution is normally applied in filtering, where the input signal $x(n)$ has length much greater than $h(n)$ (the system impulse response). To process such convolution the input signal is partitioned in segments of length $M_0$, obtaining
\[
x_m(n) = \left\{\begin{array}{c c} x(n - mM_0) & for \ 0 \leq n \leq M-1\ \text{and} \ m\ \text{integer} \\ 0, & \text{otherwise}. \end{array}\right.
\]
Considering $h(n)$ of length $L$ and $N = M_0 + L -1$ results in
\[
y_m(n) = \sum_{k=0}^{N-1}x_m(k)h(n-k),\ \text{for} \ 0 \leq n \leq N - 1,
\]
where $y_m(n)$ is the linear convolution between the $m$-th segment of $x(n)$ and $h(n)$.

Writing the input sequence as 
\[
x(n) = \sum_{m = 0}^{\infty}x_m(n - mM_0)
\]
by the superposition principle
\[
y(n) = \sum_{m=0}^{\infty}y_m(n - mM_0).
\]
However as $N = M_0 + L -1 > M$ there will be superposition between consecutive $y_m(n)$ as represented in Figure~\ref{fig:OverlappAdd}. The superpositions are added to obtain the final signal $y(n)$, thus this convolution method is named \index{Overlapp-Add Method} Overlapp-Add.

\subsection{DFT Leakage}

DFT is an approximation of the signal spectrum  representation obtained by taking $N$ samples of the continuous function $X(e^{j\omega})$ producing a $N$ point sequence $X(k)$. According to Equation~\ref{eq:relationfanalogfanalytical} when frequency components of a signal  are exactly located at $f_k$, $0 \leq k \leq N-1$, all energy of each components is represented exactly at the respective $X(k)$. For example, consider an analog signal $x(t) = \cos(2\pi 1000t)$ sampled with $f_s = 4 kHz$ producing the sampled signal $x(n) =\cos(\frac{\pi}{2} n)$. Taking a $N=16$ point DFT of $x(n)$ we obtain $X(k)$ represented in Figure~\ref{fig:leakage1}. From Equation~\ref{eq:relationfanalogfanalytical} $X(4)$ represents exactly the frequency $4\times 4000/16 = 1 kHz$ of the input signal.

An alternative view to DFT leakage is to model a $N$ points signal $x(n)$ as a multiplication between $x(n)$ and a  ``limiting" function, i.e.
\begin{equation} 
x_N(n) = x(n)w(n)
\label{eq:limitatempodiscreto}
\end{equation}
where $w(n)$ is a rectangular window signal given by
\[
w(n) = \left\{ \begin{array}{c c} 1 & 0 \leq n \leq N - 1 \\ 0 & \text{otherwise.} \end{array}\right.
\]
Signal $w(n)$ is the same as in Example~\ref{ex:portadiscreta} with DFT calculated in Equation~\ref{eq:DFTPulsoQuadradoN16} for $N=16$, 
\[
W(k) = \sum_{n=0}^{15}x(n)e^{-j(\frac{2\pi  k}{16})n} = \frac{e^{-j2\pi k} -1}{e^{-j(2\pi k/16)}-1}.
\]


Applying DFT in Equation~\ref{eq:limitatempodiscreto} implies in circularly convolving $X(k)$ with $W(k)$ what has the effect of spreading each component of $X(k)$ to the adjacent positions. 

When frequency components of $x(n)$ are exactly located at one of $k$ in $X(k)$ leakage will not occur. For that case for each sample $X(k)$ value of  $W(k)$ is maximum in $k$ and zero in adjacent samples. %, as represented in Figure~\ref{fig:sincsinchronized}.

\section{Windowing}

As shown in the previous section leakage effect results from the limited number of values taken from the sequence $x(n)$, modeled as a multiplication by a window  sequence $w(n)$. To limit the leakage effect another window function with limited energy can be used. For example using a Hanning window 
\[
w(n) = 0.5 - 0.5\cos\left(\frac{2\pi n}{N - 1}\right),\ \ n = 0, 1, 2, ..., N-1. 
\]
represented in Figure~\ref{fig:HanningWindow} a) for $N = 16$ with magnitude of the DFT represented in Figure~\ref{fig:HanningWindow} b), that have more concentrated frequency components than the rectangular function frequency components. As we consider the limiting function  $x_N(n) = x(n)w(n)$ the spectrum representation of $X_N(k)$ will be more concentrated. Multiplying the Hanning window to  $x(n) = \cos(\frac{11\pi}{20}n)$, whose absolute value of the DFT is presented in Figure~\ref{fig:leakage2}, the representation in Figure~\ref{fig:Cossinewindowed} is obtained.
